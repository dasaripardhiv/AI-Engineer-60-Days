# Smart Light Intelligent Agent using PEAS Framework

This notebook implements a simple reflex intelligent agent that turns
a light ON or OFF based on brightness and human presence.



```python
class SmartLightAgent:
    def __init__(self):
        print("Smart Light Agent Initialized")

    def perceive(self, brightness, presence):
        return brightness, presence

    def act(self, brightness, presence):
        if brightness == "dark" and presence == "yes":
            return "LIGHT ON"
        else:
            return "LIGHT OFF"
```

```python
agent = SmartLightAgent()

print("Scenario 1:", agent.act("dark", "yes"))
print("Scenario 2:", agent.act("bright", "yes"))
print("Scenario 3:", agent.act("dark", "no"))
```

```python
brightness = input("Enter brightness (bright/dark): ").lower()
presence = input("Is anyone present? (yes/no): ").lower()

result = agent.act(brightness, presence)
print("Agent Decision:", result)
```
